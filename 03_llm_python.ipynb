## Lokales Modell verwenden

```python
print(response["response"])
```

## Lokales Modell zum Automatisieren

```python
from os import listdir

for file in listdir("output"):
    with open("output/"+file, "r") as f:
        quote = f.read()

    prompt = "Interpretiere dieses Zitat: " + quote

    response = generate(LOCAL_MODEL, prompt)

    with open("interpretations/"+file, "w") as output:
        output.write(response["response"])
```
